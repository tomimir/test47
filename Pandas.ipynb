```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("AAPL.xlsx")
data
```

```python
mascara=data.timestamp >= "2008-05-15"
data[mascara].tail(n=3)
```

```python
data1=data.set_index("timestamp")
data1.loc["2008-05-20":"2008-05-15"]
data1.tail(3)
```

```python
mascara1=np.logical_and(data.timestamp >="2008-05-15",data.timestamp <="2011-08-20")
print(data[mascara1])
```

```python
mascara2=np.logical_and(np.logical_and(data.timestamp >="2008-05-15",data.timestamp <="2011-08-20"),data.close>data.open)
print(data[mascara2])
```

```python
mascara2=np.logical_and(np.logical_and(data.timestamp >="2008-05-15",data.timestamp <="2011-08-20"),data.high/data.open>1.07)
data[mascara2].drop(['volume','adjusted_close'],axis=1)
```

```python
data["up intradiario"]=data.high/data.open
mascara3=np.logical_and(data.timestamp>"2008-01-01",data.timestamp<"2008-12-31")
data[mascara3].sort_values("up intradiario",ascending=False).head()
```

```python
data["intradiario"]=data.low/data.open - 1
mascara3=np.logical_and(data.timestamp>"2018-01-01",data.timestamp<"2018-12-31")
data[mascara3].sort_values("intradiario",ascending=True).head().drop('upsiede_intradiario',axis=1)
```

```python
data["volumen"]=data.volume/data["volume"].mean()
mascara4=data.volumen>5
data[mascara4].sort_values("volumen",ascending=True).head(10)
print(data["volume"].mean())
data[mascara4]
```
