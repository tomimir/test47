```python
import numpy as np
def tirar_n_dados(n):
    tirada = []
    for n in range(n):
        tirada.append(np.random.randint(1,7))
    return np.array(tirada)
```

```python
tirar_n_dados(5)
```

```python
def es_generala(x):
    dado=np.unique(x)
    if(len(dado)==1):
        return True
    else:
        return False
    
```

```python
j=0
for i in range(10000):
    n=tirar_n_dados()
    y=es_generala(n)
    if(y==True):
        j=j+1
        print(n)
print(j)   
```

```python
def seleccionar_mayoria(x):
    unico=np.unique(x)
    freq=np.zeros(len(unico))
    for i in range(len(unico)):
        for j in range(len(x)):
            if unico[i]==x[j]:
                freq[i]+=1
    maximo=unico[np.argmax(freq)]
    frecuencia=int(freq[np.argmax(freq)])
    return maximo,frecuencia     
    
```

```python
n=tirar_n_dados()
print(seleccionar_mayoria(n))
```

```python
def jugar(n_tiradas= 3, dados_total= 5):
    n_dados = dados_total 
    dados_mesa = np.array([]) 
    tiradas = np.zeros((n_tiradas,dados_total))
    tirada_exitosa= 0
    contador_tiradas=0   
    
    while (contador_tiradas < n_tiradas) and tirada_exitosa == 0:
        dados_nuevos = tirar_n_dados(n_dados)
        todos_dados = np.concatenate ((dados_nuevos,dados_mesa))
        tiradas[contador_tiradas]= todos_dados
        
        if es_generala(todos_dados):
            tirada_exitosa = contador_tiradas+1
        else:
            maximo, frecuencia = seleccionar_mayoria (todos_dados)
            dados_mesa = np.repeat(maximo, frecuencia)
            n_dados = dados_total -len(dados_mesa) 
        contador_tiradas+=1
        
    return tiradas, tirada_exitosa
```

```python
for i in range(20000):
    tiradas, tirada_exitosa = jugar(3,5)
    if tirada_exitosa >=0:
        print(tiradas, tirada_exitosa)
```
